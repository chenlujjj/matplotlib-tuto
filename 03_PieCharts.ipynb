# Pie Charts

```python
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter

plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('data.txt')

language_counter = Counter()
for lang in data['LanguagesWorkedWith']:
    language_counter.update(lang.split(';'))


languages = []
popularity = []

for item in language_counter.most_common(5):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
explode = [0, 0, 0, 0, 0.1]

plt.pie(popularity, labels=languages, autopct='%1.1f%%',
        startangle=90, shadow=True, explode=explode,
        wedgeprops={'edgecolor': 'black'})
```

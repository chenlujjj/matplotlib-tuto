```python
from matplotlib import pyplot as plt
```

```python
# 设定 style
# 查看有哪些可选的 style
print(plt.style.available)
# plt.style.use('fivethirtyeight')
plt.style.use('ggplot')

# 还有一种玩法，将图画风格设定为 xkcd 漫画风， lol
plt.xkcd()
```

绘制折线图非常简单，只需传入两个长度相同的数字列表，然后加上一些定制参数：线型、颜色、标签等

```python
plt.plot(ages_x, dev_y, color='#FF00FF', linestyle='--', marker='.', label='All Devs')
plt.plot(ages_x, py_dev_y, color='b', linestyle='-', label='Python Devs')
plt.plot(ages_x, js_dev_y, color='g', linestyle='-', label='JavaScript Devs')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()

plt.grid(True)
plt.tight_layout()

plt.show()
```
